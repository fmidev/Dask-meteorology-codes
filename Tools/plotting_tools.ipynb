```python
%%capture
%pip install matplotlib --upgrade --user
%pip install cartopy --upgrade --user

import ipywidgets as widgets
from matplotlib import pyplot as plt
import cartopy


def get_map_layout():
    # Luodaan karttapohja
    axx = plt.axes(projection = cartopy.crs.LambertConformal(central_longitude=int(np.mean(lon)), central_latitude=int(np.mean(lat))))
    axx.gridlines()
    #axx.set_extent([-5, 40, 48, 74])
    axx.add_feature(cartopy.feature.COASTLINE)
    axx.add_feature(cartopy.feature.BORDERS)
    axx.add_feature(cartopy.feature.OCEAN)
    axx.add_feature(cartopy.feature.LAND)
    axx.figure.set_size_inches([14, 9])
    return axx


def generate_fig_properties(lat, lon):
    proj = cartopy.crs.LambertConformal(central_latitude=int(np.mean(lat)),
                                        central_longitude=int(np.mean(lon)),
                                        standard_parallels=(25, 25))
    ax = plt.axes(projection=proj)
    #ax.set_extent([-5, 40, 48, 74])
    ax.gridlines()
    ax.add_feature(cartopy.feature.COASTLINE)
    ax.add_feature(cartopy.feature.BORDERS)
    ax.add_feature(cartopy.feature.OCEAN)
    ax.add_feature(cartopy.feature.LAND)
    ax.figure.set_size_inches([14, 9])
    return ax
    

def plot_map_xarray(chart, vmin, vmax, cmap, axx):
    # Plotataan itse data
    cm = chart.plot(x="lon", y="lat", ax=axx, shading='auto', transform = cartopy.crs.PlateCarree(),
                          cmap = cmap, vmin = vmin, vmax = vmax, add_colorbar=False)
    plt.colorbar(cm, fraction=0.046, pad=0.04, orientation="vertical")
    #plt.show()



def pcolormesh_map_xarray(chart, vmin, vmax, cmap, axx):
    # Plotataan itse data
    cm = chart.plot.pcolormesh(x="lon", y="lat", ax=axx, shading='auto', transform = cartopy.crs.PlateCarree(),
                          cmap = cmap, vmin = vmin, vmax = vmax, add_colorbar=False)
    plt.colorbar(cm, fraction=0.046, pad=0.04, orientation="vertical")



def pcolormesh_map_plt(data, vmin, vmax, cmap, lon, lat):
    # Tapa kaksi, on hieman nopeampi
    cm = plt.pcolormesh(lon, lat, data, transform=cartopy.crs.PlateCarree(), 
                       shading='auto', cmap = cmap, vmin = vmin, vmax = vmax)
    plt.colorbar(cm, fraction=0.046, pad=0.04, orientation="horizontal")


def contourf_map_plt(data, vmin, vmax, cmap, lon, lat):
    # Luodaan kuva
    # Tapa on melko hidas
    levels = np.linspace(vmin, vmax, 40)
    cm = plt.contourf(lon, lat, data, 30, levels=levels, transform=cartopy.crs.PlateCarree(), cmap = cmap)
    plt.colorbar(cm, fraction=0.046, pad=0.04, orientation="horizontal")
```
